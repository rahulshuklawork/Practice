```python
# Import all the required liabrary

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load Sonar Data Set
df = pd.read_csv("C:/Users/acer/Desktop/Data Science/Data Sets/Mushroom/agaricus-lepiota.csv")
df.head()
```

```python
#check the shape of data
df.shape
```

```python
#Check if any missing value present in data
df.isnull().sum()
```

```python
df.dtypes
```

```python
#Get the count of class of Mushroom (classes: edible=e, poisonous=p)
df['Classes'].value_counts()
```

```python
#Visualize the count
sns.countplot(df['Classes'],label='count')
```

```python
#Encode the categorical data values
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
df['Classes'] = label.fit_transform(df['Classes'].values)
df
```

```python
#get dummies of other columns
df = pd.get_dummies(df)
df.head()
```

```python
#split the dataset into independent (x) and dependent (y) data sets
x = df.iloc[:,1:]
y = df['Classes']
```

```python
x
```

```python
y
```

```python
# split the data set into training (80%) and testing (20%)
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = .30 , random_state = 42)
```

```python
#Create models (Logistic Regression)

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

log = LogisticRegression()
log.fit(x_train,y_train)
logpred = log.predict(x_test)

print("accuracy_score",accuracy_score(y_test,logpred))
print(confusion_matrix(y_test,logpred))
print(classification_report(y_test,logpred))
```

```python
#see ROC Curve
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt
from sklearn.metrics import roc_auc_score

y_pred_prob = log.predict_proba(x_test)[:,1]
y_pred_prob
```

```python
fpr,tpr,thresolds=roc_curve(y_test,y_pred_prob)
```

```python
fpr
```

```python
tpr
```

```python
plt.plot([0,1],[0,1],'k--')
plt.plot(fpr,tpr,label='Logidtic Regression')
plt.xlabel('false positive rate')
plt.ylabel('true positive rate')
plt.title('Logidtic Regression')
plt.show()
```

```python
auc_score=roc_auc_score(y_test,log.predict(x_test))
print(auc_score)
```
